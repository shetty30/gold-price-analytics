```python
import pandas as pd
import yfinance as yf

tickers = {
    "gold": "GLD",
    "dxy": "DX-Y.NYB",
    "crude": "CL=F",
    "us10y": "^TNX",
    "sp500": "^GSPC"
}

start_date = "2021-01-01"   
end_date = None            

data = {}

for name, ticker in tickers.items():
    df = yf.download(ticker, start=start_date, end=end_date)
    data[name] = df["Close"]


final_df = pd.concat(data, axis=1)


final_df = final_df.dropna()


final_df.reset_index(inplace=True)
final_df.to_csv("gold_macro_5y.csv", index=False)

final_df.head()
```

```python
import pandas as pd

df = pd.read_csv("gold_macro_5y.csv")
df.head()
```

```python
df.shape , df.columns
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")
df.info()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df["Date"], df["gold"])
plt.title("Gold Price â€“ Last 5 Years")
plt.show()
```

```python
df.describe
```

```python
import numpy as np


df["gold_return"] = np.log(df["gold"] / df["gold"].shift(1))

df = df.dropna()

df[["Date","gold","gold_return"]].head()
```

```python
df["gold_vol_30d"] = df["gold_return"].rolling(30).std() * np.sqrt(252)

df[["Date","gold_vol_30d"]].tail()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,1, figsize=(12,8), sharex=True)

ax[0].plot(df["Date"], df["gold"])
ax[0].set_title("Gold Price")

ax[1].plot(df["Date"], df["gold_vol_30d"])
ax[1].set_title("30 Day Rolling Volatility")

plt.tight_layout()
plt.show()
```

```python
df.to_csv("gold_macro_features.csv", index=False)
```
